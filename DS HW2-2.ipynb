```python
import requests
url = 'https://huggingface.co/models'
html = requests.get(url)
html.encoding = "utf-8"

from bs4 import BeautifulSoup
sp = BeautifulSoup (html.text, 'lxml')
```

```python
row = sp.find_all("article")
```

```python
at = []
for i in row:
    no_t = i.text.replace('\t', '@')
    no_n = no_t.replace('\n', '@')
    no_d = no_n.replace('â€¢', '@')
    no_d = no_d.split('@')
    temp = []
    for i in range(len(no_d)):
        if no_d[i] != '' and no_d[i] != ' ':
            temp.append(no_d[i])
    at.append(temp)
```

```python
done = []
for i in at:
    if len(i) == 4:
        dic = {
            'Project': i[0],
            'Type': None,
            'Update': i[2],
            'Downloads': None,
            'Likes': i[3]
        }
        done.append(dic)
    elif len(i) == 5:
        dic = {
            'Project': i[0],
            'Type': i[1],
            'Update': i[3],
            'Downloads': None,
            'Likes': i[4]
        }
        done.append(dic)
    else:
        dic = {
            'Project': i[0],
            'Type': i[1],
            'Update': i[3],
            'Downloads': i[4],
            'Likes': i[5]
        }
        done.append(dic)
```
